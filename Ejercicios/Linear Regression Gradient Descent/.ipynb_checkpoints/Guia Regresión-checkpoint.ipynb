# Guia regresión 
## Consigna
Para el dataset a continuación, calcular la regresión polinómica e ir variando el orden hasta obtener el resultado óptimo de AIC y BIC. También calcular $ R^2 $ y los p values.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.metrics import explained_variance_score

X = np.load('X.npy')
y = np.load('y.npy')
plt.figure()
plt.plot(X, y,'.',  color='blue')
plt.title('Dataset')
plt.show()


```

## Solución

```python
# create X and y

#X = X.reshape(-1,1)
#y = y.reshape(-1,1)
#X=np.reshape(X,(X.shape[0],1))

# armo un dataframe con los datos# armo  
df = pd.DataFrame({'indpte':X, 'dpte':y})
df.shape
```

```python

# follow the usual sklearn pattern: import, instantiate, fit
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor

#lm = SGDRegressor()
lm = LinearRegression()
#lm.fit(X, y)
Estimadores = smf.ols(formula='dpte ~ indpte', data=df).fit()

# print intercept and coefficients
# print the coefficients
Estimadores.params
print(Estimadores.summary())
```

```python
# Agrego el termino constante
Xc = sm.add_constant(X)

#Ajuste el modelo 
Estimadores = sm.OLS(y, Xc).fit()

#Parámetros del modelo (bo-b1)
Estimadores.params

#Resumen Minimos cuadrados ordinarios 
print(Estimadores.summary())
```

#### Los estimadores b0 y b1 son ambos significativos p-value<5%
#### R2= 0,56 y R2 Adj=0,56. 
#### El modelo es significativo Prob. F-Stat <5%

#### Cambio el grado del polinomio para testear posibles mejores ajustes ya que el gráfico no termina de ajustar con precisión.


```python
# agrego las potencias de los x para hacer las regresiones polinómicas
N = 10 # orden máxima de polinomios a probar
df_aux = pd.DataFrame(columns = ['indpte1'])
df_aux['indpte1'] = df['indpte']
for i in range(2,N+1):
    indice = 'indpte' + str(i)
    tmp_df = pd.DataFrame(columns = [indice])
    tmp_df[indice] = df['indpte']**i
    df_aux = pd.concat([df_aux,tmp_df], axis=1)

df_aux = pd.concat([df_aux,df['dpte']], axis=1)

# separo el df en entrenamiento y prueba (70 % para entrenamiento)
msk = np.random.rand(len(df)) < 0.7

df_train = df_aux[msk]
df_test = df_aux[~msk]
```

```python
df_aux.head 
```

```python
# Testeo diferentes grados
resultados = pd.DataFrame(columns=['Grado','AIC','BIC','R2'])

for N in range (1,6):
    Xn = df_train.iloc[:,0:N]
    Xn = Xn.iloc[:,:].values
    y1 = df_train['dpte']

    #modelo
    Xn = sm.add_constant(Xn)
    model = sm.OLS(y1, Xn).fit()

    # Resultados
    print('=================================================================================')    
    plt.figure()    
    plt.plot(df_train['indpte1'], df_train['dpte'],'.',  color='blue')
    plt.plot(df_train['indpte1'], model.predict(), c='red', linewidth=2)
    plt.title('Orden '+ str(N))
    plt.show()
    
    resultados = resultados.append({'Grado':N,'AIC':model.aic,'BIC':model.bic,'R2':model.rsquared},ignore_index=True)
```

```python
print(resultados)
```

```python
N  =  3
Xn = df_train.iloc[:,0:N]
Xn = Xn.iloc[:,:].values
y1 = df_train['dpte']
```

```python
#modelo
Xn = sm.add_constant(Xn)
model = sm.OLS(y1, Xn).fit()


# Resultados
plt.figure()
plt.plot(df_train['indpte1'], df_train['dpte'],'.',  color='blue')
plt.plot(df_train['indpte1'], model.predict(), c='red', linewidth=2)
plt.title('Train: Orden '+ str(N))
plt.show()

print(model.summary())
print()
print('PValues:')
print(model.pvalues)

# Calculo el score del set de entrenamiento
print()
print('Score Train:')
print(explained_variance_score(df_train['dpte'], model.predict(Xn)))
```

```python
#### Modelo sin Intercepto(no agrego la constante) ni x1. 
N  =  3
Xn = df_train.iloc[:,1:N]
Xn = Xn.iloc[:,:].values
y1 = df_train['dpte']

#modelo
#Xn = sm.add_constant(Xn)
model = sm.OLS(y1, Xn).fit()

# Resultados
plt.figure()
plt.plot(df_train['indpte1'], df_train['dpte'],'.',  color='blue')
plt.plot(df_train['indpte1'], model.predict(), c='red', linewidth=2)
plt.title('Train: Orden '+ str(N))
plt.show()

print(model.summary())
print()
print('PValues:')
print(model.pvalues)

# Calculo el score del set de entrenamiento
print()
print('Score Train:')
print(explained_variance_score(df_train['dpte'], model.predict(Xn)))
```

```python
# Veo como performa con los datos de testing# Veo c 
Xn_test = df_test.iloc[:,1:N]
Xn_test = Xn_test.iloc[:,:].values
y1_test = df_test['dpte']

# Resultados
plt.figure()
plt.plot(df_test['indpte1'], df_test['dpte'],'.',  color='blue')
plt.plot(df_test['indpte1'], model.predict(Xn_test), c='red', linewidth=2)
plt.title('Test: Orden '+ str(N))
plt.show()
```

```python
# Calculo el score del set de prueba
print('Score Test:')
print(explained_variance_score(df_test['dpte'], model.predict(Xn_test)))
```
